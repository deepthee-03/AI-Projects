```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
from scipy import integrate

# Given parameters
mean = 170  # Mean height (cm)
std = 7  # Standard deviation of height (cm)

# Generate x values for the plot
x = np.linspace(mean - 3 * std, mean + 3 * std, 1000)

# Calculate the PDF for the normal distribution
pdf = norm.pdf(x, loc=mean, scale=std)

# Define the range of values (heights) you're interested in
range_start = 160
range_end = 180

# Integrate the PDF over the specified range to calculate the probability within the range
probability, _ = integrate.quad(norm.pdf, range_start, range_end, args=(mean, std))

# Mask the PDF values outside the specified range
pdf_masked = np.where((x >= range_start) & (x <= range_end), pdf, 0)

# Plot the PDF with shading for the specified range
plt.figure(figsize=(9, 8))
plt.plot(x, pdf, label='PDF (Normal)')
plt.fill_between(x, pdf_masked, alpha=0.2, color='blue')
plt.axvline(x=range_start, color='green', linestyle='--', label='Range Start')
plt.axvline(x=range_end, color='red', linestyle='--', label='Range End')
plt.xlabel('Height (cm)')
plt.ylabel('Probability Density')
plt.title('Probability Density Function (PDF) of Normal Distribution within Range')
plt.legend()
plt.grid(True)
plt.show()

print(f"Probability within range [{range_start}, {range_end}]: {probability:.4f}")
```
