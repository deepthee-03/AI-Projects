```python
import random
print(random.randint(0,10))
from numpy import random
x=random.normal(size=(2,3))
print(x)
import matplotlib.pyplot as plt
import seaborn as sns
sns.distplot(random.normal(size=1000), hist=True,label='normal')
plt.title('Continous(Normal Distribution)', fontsize='20')
plt.xlabel('X Axis', fontsize='20')
plt.ylabel('Y Axis', fontsize='20')
plt.legend()
plt.show()
```

```python
import random
print(random.randint(0,10))
from numpy import random
x = random.uniform(size=(2, 5))
print(x)
import matplotlib.pyplot as plt
import seaborn as sns
sns.distplot(random.uniform(size=1000), hist=True,label='uniform')
plt.title('Continous(Uniform Distribution)', fontsize='20')
plt.xlabel('X Axis', fontsize='20')
plt.ylabel('Y Axis', fontsize='20')
plt.legend()
plt.show()
```

```python
import random
print(random.randint(0,10))
from numpy import random
import matplotlib.pyplot as plt
import seaborn as sns
sns.distplot(random.binomial(n=10, p=0.5, size=1000), hist=True,
kde=True,label='Binomial')
plt.title('Discrete(Binomial Distribution)', fontsize='15')
plt.xlabel('X Axis', fontsize='10')
plt.ylabel('Y Axis', fontsize='10')
plt.legend()
plt.show()
```

```python
import random
print(random.randint(0,10))
from scipy.stats import bernoulli
import matplotlib.pyplot as plt
bd=bernoulli(0.8)
x=[0,1]
plt.figure(figsize=(5,5))
plt.xlim(-3, 2)
plt.bar(x, bd.pmf(x), color='blue')
plt.title('Discrete(Bernoulli distribution)(p=0.8)', fontsize='15')
plt.xlabel('Values of random variable x (0, 1)', fontsize='10')
plt.ylabel('Probability', fontsize='10')
plt.legend()
plt.show()
```
