```python
import numpy as np
import matplotlib.pyplot as plt

# Generate random values for sensitivity, base rate, and likelihood
sensitivity = np.random.uniform(0.8, 1.0)  # Random sensitivity between 0.8 and 1.0
base_rate = np.random.uniform(0.01, 0.10)  # Random base rate between 0.01 and 0.10 (low to represent a rare disease)
likelihood = np.random.uniform(0.7, 0.95)  # Random likelihood between 0.7 and 0.95

# Define a fixed prior probability
prior_probability = 0.7

# Calculate posterior probability using Bayes' theorem
posterior = (sensitivity * prior_probability) / (
    (sensitivity * prior_probability) + (1 - base_rate) * (1 - prior_probability)
)

# Plotting
plt.bar(['Prior', 'Posterior'], [prior_probability, posterior], label='Prior')
plt.ylabel('Probability')
plt.title('Prior and Posterior Probabilities for Disease Testing')
plt.ylim(0, 1)
plt.legend(['Posterior'], prop={'size': 10})
plt.show()

# Print the generated values
print(f"Generated Sensitivity: {sensitivity:.2f}")
print(f"Generated Base Rate: {base_rate:.2f}")
print(f"Generated Likelihood: {likelihood:.2f}")
print(f"Calculated Posterior Probability: {posterior:.4f}")
```
